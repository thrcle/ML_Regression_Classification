```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
data=pd.read_csv('/content/drive/MyDrive/패캠/data/data_2020.csv')
data['거리']=data['도착영업소코드'].map({105: 20, 110: 77, 115: 135, 120: 185, 125: 240, 130: 215, 135: 339, 140: 407}) 
data.pop('도착영업소코드')
data
```

```python
data_destination=data[data['거리']==407]  # 그곳은 부산
data_destination
```

```python
df=data_destination.groupby(['집계시','요일','거리'])['통행시간'].mean()
df
```

```python
df2=df.unstack().dropna().reset_index()
df2
```

```python
stat=df2.describe()
stat
```

```python
value_25=stat[407]['25%']  # = value_25=stat[407][4]
value_75=stat[407]['75%']
```

```python
df2['Grade']=0

def get_grade(val):
  if val>value_25:
    if val>value_75:
      output=2
    else:
      output=1
  else:
    output=0
  return output

  
```

```python
# df2['grade']=get_grade(df2[407])
```

```python
data_grade=pd.DataFrame(columns=['time','day','grade'])

for index, row in df2.iterrows():                                       # iterrows()
  print(index, row['집계시'], row['요일'])
  data_grade = data_grade.append({
      'time':row['집계시'],
      'day':row['요일'],
      'grade':get_grade(row[407])
  }, ignore_index=True)

```

```python
data_list=data_grade.values.tolist()
data_list[:5]
```

```python
train_x=[r[0:2] for r in data_list]
train_y=[[r[-1]] for r in data_list]

#y(grade) One hot encode [0,1,2] to [[1,0,0,]. [0,1,0],[0,0,1]]
y_one_hot=tf.keras.utils.to_categorical(train_y)
```

```python
# Tensorflow
learning_rate = 1e-2
learning_epochs = 1000

# Stochastic gradient descent (SGD) Optimizer
sgd = tf.keras.optimizers.SGD(learning_rate=learning_rate)

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(3, input_shape=(2,), activation='softmax'))   # output이 복수개니까 softmax. (1개라면 sigmoid ok)

model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_x
```

```python
import numpy as np

history=model.fit(np.array(train_x), np.array(y_one_hot), epochs=learning_epochs)    # 이 경우 왜 array로?
```

```python
model.summary(train_x, train_y)
```

```python
time=14
day=3

# 실제값
print( df2[ (df2['집계시']==time) & (df2['요일']==day) ]  )

# 예측값

result = model.predict(np.array([ [time,day] ])) 
print(result)

grade_list = ['빠름', '보통', '느림']
grade_index=np.argmax(result[0])  #argmax : 가장 큰 값을 취함
grade=grade_list[grade_index]
```
